# Earthquakes Jul / Aug 2016

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#Pull the data in from the csv and parse the time.full column
eq = pd.read_csv("earthquakes.csv", parse_dates=["time.full"])
```

```python
#Seems the pyplot requires a DataFrame, a group by on single column produces a Series, convert using pandas to a DataFrame
eqGrpDf = pd.DataFrame(eq.groupby(["location.name"])["impact.magnitude"].agg("max").nlargest(n=20))
```

```python
#The DataFrame need the index resetting so the key becomes first column in this case location.name
eqGrpDf.reset_index(inplace=True)
```

## Who had the largest magnitude earthquakes that month?

```python
eqGrpDf.head(20)
```

Let's plot these results

```python
#Using the plyplot oo method, needs the DataFrame version of the data, and a show or savefig
x = eqGrpDf["location.name"]
y = eqGrpDf["impact.magnitude"]


colors = ['salmon'] #, 'green', 'blue', 'yellow', 'black', 'cyan']

plt.figure(figsize=(20,5))
plt.grid(True)
plt.bar(x, y, width = 0.5, color = colors)
plt.xlabel("Locations", fontsize = 15)
plt.ylabel("Magnitude", fontsize  = 15)
plt.title("Earthquakes Jul/Aug 2016", fontsize  = 20)
plt.xticks(rotation = 90,  fontsize = 15)
plt.yticks(rotation = 0,  fontsize = 15)


plt.show()
```

## Who had the most earthquakes that month ?

```python
eqGrpDf = pd.DataFrame(eq.groupby(["location.name"])["impact.magnitude"].agg("count").nlargest(n=20))
```

```python
eqGrpDf.head(20)
```

Let's plot these results

```python
#The DataFrame need the index resetting so the key becomes first column in this case location.name
eqGrpDf.reset_index(inplace=True)
```

```python
#Using the plyplot oo method, needs the DataFrame version of the data, and a show or savefig
x = eqGrpDf["location.name"]
y = eqGrpDf["impact.magnitude"]


colors = ['salmon'] #, 'green', 'blue', 'yellow', 'black', 'cyan']

plt.figure(figsize=(20,5))
plt.grid(True)
plt.bar(x, y, width = 0.5, color = colors)
plt.xlabel("Locations", fontsize = 15)
plt.ylabel("Count", fontsize  = 15)
plt.title("Earthquakes Jul/Aug 2016", fontsize  = 20)
plt.xticks(rotation = 90,  fontsize = 15)
plt.yticks(rotation = 0,  fontsize = 15)


plt.show()
```

## Lets look at California over time

```python
locationfilter = (eq["location.name"] == "California")
eqTime = eq.loc[locationfilter]
```

```python
#Using the plyplot oo method, needs the DataFrame version of the data, and a show or savefig
x = eqTime["time.full"]
y = eqTime["impact.magnitude"]
colors = ['red' ] #, 'green'] #, 'blue', 'yellow', 'black', 'cyan']

plt.figure(figsize=(20,10))
plt.grid(True)
plt.scatter(x, y, color = colors)
plt.xlabel("time", fontsize = 15)
plt.ylabel("magnitude", fontsize  = 15)
plt.title("California Earthquakes Jul/Aug 2016", fontsize  = 20)
plt.xticks(rotation = 90,  fontsize = 15)
plt.yticks(rotation = 0,  fontsize = 15)
plt.show()
```
