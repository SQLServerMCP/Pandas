<h1>Exoplanet Discovery Data
<h3>Distribution of exoplanet discoveries

Analysis of data on the exoplanet discoveries up to 2016 from a data set downloaded from <br>
https://exoplanetarchive.ipac.caltech.edu/cgi-bin/TblView/nph-tblView?app=ExoTbls&config=PSCompPars


```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
exo = pd.read_csv("ExoPlanetData.csv", skiprows = 173)
```

```python
exo.set_index("rowid")
```

```python
##Only want these columns
##rowid	pl_name	hostname	sy_snum	sy_pnum	discoverymethod	disc_year	disc_locale	
##disc_telescope	pl_radj	 pl_rade  pl_bmasse	pl_bmassj
```

<h2>What is the distribution by discovery method

```python
exo["discoverymethod"].value_counts()
```

```python
exo["discoverymethod"].value_counts().plot(kind='bar', title = 'Exoplanets by Discovery Method', color = 'gray',
                grid=True, xlabel = 'Discovery Method', ylabel = 'Exoplanets', figsize=(20,7), fontsize = 15)
```

<h2>Distribution by Year Discovery

```python
exo.groupby(["disc_year"])["disc_year"].agg('count').plot(kind ='bar', title = 'Exoplanets by Year Discovery'
,grid=True, color='gray',xlabel = 'Year of Discovery', ylabel = 'Exoplanets', figsize=(20,7), fontsize = 15)
```

<h2>Distribution by Telescope (Top 10)

```python
exo["disc_telescope"].value_counts().head(10).plot(kind = 'bar', title = 'Exoplanets by Telescope',grid=True,
    color='gray',xlabel = 'Telescope', ylabel = 'Exoplanets', figsize=(20,7), fontsize = 15)
```
